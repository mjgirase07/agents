```python

from dotenv import load_dotenv
from openai import OpenAI
import json
import os
import requests
from pypdf import PdfReader
import gradio as gr
```

```python
load_dotenv(override=True)
```

```python
reader = PdfReader("me/Mahendra_CV.pdf")
mahendra_cv=""
for page in reader.pages:
    text = page.extract_text()
    if text:
        mahendra_cv += text

# with open("me/mahendra_cv.pdf","r",encoding="utf-8") as f:
#     summary = f.read()
summary=mahendra_cv
name = "Mahendra Girase"
```

```python
system_prompt = f"You are acting as {name}. You are answering questions on {name}'s website, \ particularly questions related to {name}'s career, background, skills ans experience. \ Your responsibility is to represent {name} for interactions on the website as faithfully as possible. \ You are given a summary of {name}'s background and Mahendra CV(resume) which you can use to answer questions. \ Be professional and engaging, as if talking to a potential client or future employer who came across the website. \ If you don't know the answer to any question tell humbly you dont know. \ If the user is engaging in discussion, try to steer them towards getting in discussion, try to steer them towards getting in touch via email; provide them Mahendra email mention."

system_prompt += f"\n\n## Summary:\n{summary}\n\n## Mahendra CV:\n{mahendra_cv}\n\n"
system_prompt += f"With this context, please chat with the user, always staying in character as {name}."
```

```python
openai = OpenAI(
    base_url=os.getenv("OPENAI_BASE_URL"),
    api_key=os.getenv("OPENAI_API_KEY")
)
```

```python
def chat(message, history):
    messages = [{"role":"system", "content":system_prompt}] + history + [{"role":"user", "content":message}]
   
    #this is the call to the LLM
    response = openai.chat.completions.create(model="openai.o4-mini", messages=messages)

    return response.choices[0].message.content
```

```python
gr.ChatInterface(chat).launch()
```
